# QuickSort

```python
import matplotlib.pyplot as plt
```

```python
xQuickSort = [1000, 10000, 100000, 150000, 200000, 500000, 750000, 1000000]
yQuickSort = [0.00011900, 0.00121400, 0.02086300, 0.02591500, 0.03945100, 0.18421300, 0.38355700, 0.66949000]

plt.plot(xQuickSort, yQuickSort, label="Quick Sort")
plt.legend()
plt.ylabel("Süre (sn)")
plt.xlabel("Eleman Sayısı")
```
